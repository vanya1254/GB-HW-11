```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return -12 * x**4 * np.sin(np.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
```

```python
# так как не указан диапозон, на котором исследуем функцию, выберу свой
# x = np.arange(-2, 2, 0.01)
# graph_f = plt.plot(x, f(x), 'r-')

# корень полученный в симпае и в моем диапозоне ~
root_1 = -1.33 #-(5*sqrt(218931)/108 + 127045/5832)**(1/3)/3 - 565/(972*(5*sqrt(218931)/108 + 127045/5832)**(1/3)) + 5/54
plt.plot(root_1, f(root_1), 'go', label = 'корень')

# диапозоны возрастания
neg_x_up_1 = np.arange(-0.5, 0.5, 0.01)
neg_x_up_2 = np.arange(1.75, 2, 0.01)

# диапозоны убывания
pos_x_down = np.arange(-2, root_1, 0.01)
neg_x_down_1 = np.arange(root_1, -0.5, 0.01)
neg_x_down_2 = np.arange(0.5, 1.75, 0.01)

# добавим к графику
plt.plot(pos_x_down, f(pos_x_down), 'r-', label = 'убывает f>0')
plt.plot(neg_x_down_1, f(neg_x_down_1), 'r-.', label = 'убывает f<0')
plt.plot(neg_x_up_1, f(neg_x_up_1), 'b-.', label = 'возрастает f<0')
plt.plot(neg_x_down_2, f(neg_x_down_2), 'r-.', label = 'убывает f<0')
plt.plot(neg_x_up_2, f(neg_x_up_2), 'b-.', label = 'возрастает f<0')


plt.legend()
plt.grid()
```
